```python
# import pandas library as pd
import pandas as pd
from nuscenes.nuscenes import NuScenes
import numpy as np
import math
```

```python
nusc = NuScenes(version='v1.0-mini', dataroot='data/sets/nuscenes', verbose=True)
```

Collect scene names for each scene in mini dataset. Will do this on the big data set after.

```python
scenes = []
descriptions = []
locations = []
vehicles = []
scene_lens = []
nbr_samples = []

for i in range(len(nusc.scene)):
    # Get scene names
    my_scene = nusc.scene[i]['name']
    scenes.append(my_scene)
    
    # Get scene descriptions
    description = nusc.scene[i]['description']
    descriptions.append(description)
    
    # Get scene locations
    token = nusc.scene[i]['log_token']
    location = nusc.get('log', token)['location']
    locations.append(location)
    
    # Get scene vehicles
    vehicle = nusc.get('log', token)['vehicle']
    vehicles.append(vehicle)
    
#     # Get scene lengths
#     length = nusc.scene[i]['description']
#     nb_lens.append(length)
    
    # Get scene lengths
    sample = nusc.scene[i]['nbr_samples']
    nbr_samples.append(sample)
```

```python
# create an Empty DataFrame object
df = pd.DataFrame()
df['Scene Name'] = scenes
df['Description'] = descriptions
df['Location'] = locations
df['Vehicle Number'] = vehicles # Two Renault Zoes, one in Boston, one in Singapore
df['Nbr samples'] = nbr_samples
df['Score'] = None
df
```
